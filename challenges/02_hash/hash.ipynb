# Hashing

```python
import hashlib


def compute_sha256(file_path):
    """Compute the SHA-256 hash of a file."""
    sha256_hash = hashlib.sha256()
    with open(file_path, "rb") as f:
        # Read and update hash string value in blocks of 4K
        for byte_block in iter(lambda: f.read(4096), b""):
            sha256_hash.update(byte_block)
    return sha256_hash.hexdigest()


file_path = "../pyproject.toml"
print(compute_sha256(file_path))
```

## How does hashing verify data integrity?

If I publish an application/executable and I share the hash of the latest change, then if someone changes something afterwards (maybe malicious) then the hash is different.

A good example for something like this are github releases. Every release gets marked with the latest commit message. Meaning with this you can prevent from malicious changes after the release.
